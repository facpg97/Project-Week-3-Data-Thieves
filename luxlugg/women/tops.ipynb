```python
top_url_list = ['https://www.farfecth.com' + top.find('a').get('href') for top in tops]
final_tops = []

for link in top_url_list[0:13]:
    print(link)
    time.sleep(random.randint(5,9))
    try:
        response = requests.get(url=link, headers=user_agent)
        soup = BeautifulSoup(response.content)
        result = soup.find('p', attrs={'class':'ltr-13ze6d5-Body efhm1m90'}) 
        name = result.text
        price_result = soup.find('p', attrs={'class':'ltr-194u1uv-Heading e54eo9p0'}) 
        price = price_result.text
        highlights_result = soup.find('ul', attrs={'class': '_fdc1e5'})
        highlights = highlights_result.text
        final_tops.append({'name': name, 'price':price, 'details':highlights,'url':link})
    except:
        print('this link is bad,', link)

    
```

```python
final_tops
```

```python
tops_df = pd.DataFrame(final_tops)
tops_df
```

```python
tops_df.to_csv('women_tops.csv')
```
