```python
urltap_list = ["https://www.flytap.com" + destination.find("a").get("href") for destination in destinations_europe]
urltap_list

description_list = []

user_agent = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}

for link in urltap_list:
    time.sleep(2.3)
    try:
        response = requests.get(url=link, headers= user_agent)
        print(response.status_code)
        soup = BeautifulSoup(response.content)

        #  description 
        result = soup.find("div", attrs={'class': 'richtext-tap'})
        description = result.text

        # # price 
        # result = soup.find("div", attrs={"class":"bigger"})
        # price = result.text

        description_list.append({'description': description})

    except:
        print("Not a good link", link)
```

```python
urltap_list = ["https://www.flytap.com" + destination.find("a").get("href") for destination in destinations_europe]
urltap_list

price_list = []

user_agent = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}

for link in urltap_list:
    time.sleep(2)
    try:
            response = requests.get(url=link, headers= user_agent)
            print(response.status_code)
            soup = BeautifulSoup(response.content)

            # #  description 
            # result = soup.find("div", attrs={'class': 'richtext-tap'})
            # description = result.text

            # # price 
            result = soup.find("div", attrs={"class":"price-info"})
            price = result.text
            print(price)

            price_list.append({'price': price})

    except:
        print("Not a good link", link)
```
