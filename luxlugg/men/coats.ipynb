```python
coat_url_list = ['https://www.farfecth.com' + coat.find('a').get('href') for coat in coats]
final_coats = []

for link in coat_url_list[0:12]:
    print(link)
    time.sleep(random.randint(5,9))
    try:
        response = requests.get(url=link, headers=user_agent)
        soup = BeautifulSoup(response.content)
        result = soup.find('p', attrs={'class':'ltr-4y8w0i-Body e1s5vycj0'})
        name = result.text
        price_result = soup.find('p', attrs={'class':'ltr-194u1uv-Heading e54eo9p0'})
        price = price_result.text
        highlights_result = soup.find('ul', attrs={'class': '_fdc1e5'})
        highlights = highlights_result.text
        final_coats.append({'name': name, 'price':price, 'details':highlights,'url':link})
    except:
        print('this link is bad,', link)

    
```

```python
final_coats
```

```python
coats_df = pd.DataFrame(final_coats)
coats_df
```

```python
coats_df.to_csv('men_coats.csv')
```
